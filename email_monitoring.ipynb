```python

import email.header
from bs4 import BeautifulSoup
import mailparser

def decode_mime_words(s):
    return u''.join(
        word.decode(encoding or 'utf8') if isinstance(word, bytes) else word
        for word, encoding in email.header.decode_header(s))
```

```python
data = mail.search(None, 'ALL')
mail_ids = data[1]
id_list = mail_ids[0].split()   
first_email_id = int(id_list[0])
latest_email_id = int(id_list[-1])

for i in range(latest_email_id,latest_email_id-1, -1):
    data = mail.fetch(str(i), '(RFC822)' )
    for response_part in data:
        
        arr = response_part[0]
        # print(arr)
        if isinstance(arr, tuple):
            msg = email.message_from_string(str(arr[1],'latin-1'))
            # print(msg)
            
            email_subject = decode_mime_words(msg['subject'])
            email_from = decode_mime_words(msg['from'])
            email_received = decode_mime_words(msg['Received'].split(";")[1].strip())
            email_content = decode_mime_words(msg['MIME-Version'])
            # email_content = BeautifulSoup(msg, "html.parser").get_text()
            email_received = dateutil.parser.parse(email_received).astimezone()
            # print('From : ' + email_from + '\n')
            # print('Subject : ' + email_subject + '\n')
            # print(f'Received : {email_received}\n')
            # print("-----------------------------------")
            print('Content : ' + email_content + '\n')
```

```python
data = mail.search(None, 'ALL')
mail_ids = data[1]
id_list = mail_ids[0].split()   
first_email_id = int(id_list[0])
latest_email_id = int(id_list[-1])

for i in range(latest_email_id,latest_email_id-1, -1):
    data = mail.fetch(str(i), '(RFC822)' )
    for response_part in data:
        
        arr = response_part[0]
        # print(arr)
        if isinstance(arr, tuple):
            email_msg = mailparser.parse_from_bytes(arr[1])
            email_subject = email_msg.subject
            email_from = " ".join(email_msg.from_[0])
            email_received = email_msg.received[0]['date']
            email_content = BeautifulSoup(email_msg.body, "html.parser").get_text().strip()
            print('From : ' + email_from + '\n')
            print('Subject : ' + email_subject + '\n')
            print(f'Received : {email_received}\n')
            print("-----------------------------------")
            print('Content : ' + email_content + '\n')
            
```
